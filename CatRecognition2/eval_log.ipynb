```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
log_file = 'logs/20250107204353/logs.csv'
```

```python
df = pd.read_csv(log_file)
acc = df['accuracy'].values
loss = df['loss'].values
```

```python
# acc: mean each 10 epochs
acc = np.array([np.mean(acc[i:i+10]) for i in range(0, len(acc), 10)])

# loss: mean each 10 epochs
loss = np.array([np.mean(loss[i:i+10]) for i in range(0, len(loss), 10)])
```

```python
len(acc)
```

```python
def plot_line(line, title):
    plt.plot(line)
    plt.title(title)
    plt.ylabel('Value')
    plt.xlabel('Epoch')
    plt.show()
    
```

```python
plot_line(acc[10:], 'Accuracy')
plot_line(loss[10:], 'Loss')
```
